```python
from Bio import SeqIO, SearchIO
from Bio.Blast import NCBIWWW
```

```python
import os

folder = "GCA_000005845.2"
print(os.listdir(folder))
```

```python
def GC_CONTENT_CALCULATE(sequences):
    G = sequences.count("G")
    C = sequences.count("C")
    A = sequences.count("A")
    T = sequences.count("T")

    total = G + C + A + T

    if total == 0:
        formula = 0
    else:
        formula = round((G + C) / total * 100, 2)

    return formula

```

```python
for record in SeqIO.parse("GCA_000005845.2/cds_from_genomic.fna", "fasta"):
    seq = str(record.seq)
    gc = GC_CONTENT_CALCULATE(seq)
    print(f"{record.id} → GC%: {gc}")


    
```
